1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('previsao_de_renda.csv')
df = df[['posse_de_veiculo', 'posse_de_imovel', 'renda']]
````

````python
ax1 = sns.pointplot(data=df, x='posse_de_veiculo', y='renda')
````

````python
ax2 = sns.pointplot(data=df, x='posse_de_imovel', y='renda')
````

````python
ax3 = sns.pointplot(data=df, x="posse_de_veiculo", y="renda", hue='posse_de_imovel', dodge=True,
                   capsize=.4)
````

As duas variáveis são boas preditoras de renda. A variável posse_de_veiculo, possui uma discrepância ainda maior, sendo essa uma variável preditora melhor do que posse_de_imovel.
